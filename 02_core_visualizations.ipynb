## 🌌 CryptoStock_Oracle: Phase 2 — Multi-Asset Dynamic Analytics Dashboard
#### ---
### This module acts as our visual exploration layer. It extracts the structural data matrix cached on Day 1 from localized storage files and implements a highly detailed, 8-part data visualization canvas using **Plotly**. 
### Our primary architectural objective is to map behavioral and liquidity deviations across **decentralized protocols ($BTC$)** and **traditional public companies ($AAPL$)**.

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

print("🏁 [SYSTEM CONFIGURATION] Plotly Core Graphics Canvas Pipeline Initialized.")

# Load localized checkpoints directly into working memory
crypto_df = pd.read_csv("crypto_processed.csv", parse_dates=['Date'], index_col='Date')
stock_df = pd.read_csv("stock_processed.csv", parse_dates=['Date'], index_col='Date')

print(f"📊 Tracking logs mounted. Crypto records: {len(crypto_df)} | Equity records: {len(stock_df)}")
```

### 📈 2.1 **Structural Price Horizons & Bollinger Variance Channels**
#### We overlay the raw pricing strings with automated 20-day statistical deviation upper limits and lower floors to isolate tracking breakout regimes across both asset domains.

```python
fig_compare = make_subplots(rows=2, cols=1, shared_xaxes=False,
                          subplot_titles=('1. Bitcoin ($BTC) Market Horizon & Bollinger Envelopes', 
                                          '2. Apple ($AAPL) Equity Horizon Liquidity Spectrum'))

# Subplot 1: Bitcoin Close Price + Bollinger Envelopes
fig_compare.add_trace(go.Scatter(x=crypto_df.index, y=crypto_df['Close'], name='BTC Close', line=dict(color='#f59e0b', width=2)), row=1, col=1)
fig_compare.add_trace(go.Scatter(x=crypto_df.index, y=crypto_df['BB_Upper'], name='BTC BB Upper', line=dict(color='#f43f5e', width=1, dash='dot')), row=1, col=1)
fig_compare.add_trace(go.Scatter(x=crypto_df.index, y=crypto_df['BB_Lower'], name='BTC BB Lower', line=dict(color='#10b981', width=1, dash='dot')), row=1, col=1)

# Subplot 2: Apple Close Price + Translucent Liquidity Area
fig_compare.add_trace(go.Scatter(x=stock_df.index, y=stock_df['Close'], fill='tozeroy', fillcolor='rgba(0, 255, 255, 0.08)', name='AAPL Area', line=dict(color='#00ffff', width=2)), row=2, col=1)

fig_compare.update_layout(height=700, template='plotly_dark', title_text="🍎 vs ⚡ UNIFIED MULTI-ASSET INSIGHT TERMINAL", showlegend=True)
fig_compare.show()
```

### 📈 2.1 Charts 1 & 2: Structural Price Horizons & Bollinger Variance Channels
We overlay the raw pricing strings with automated 20-day statistical deviation upper limits and lower floors to isolate tracking breakout regimes across both asset domains.

```python
fig_compare = make_subplots(rows=2, cols=1, shared_xaxes=False,
                          subplot_titles=('1. Bitcoin ($BTC) Market Horizon & Bollinger Envelopes', 
                                          '2. Apple ($AAPL) Equity Horizon Liquidity Spectrum'))

# Subplot 1: Bitcoin Close Price + Bollinger Envelopes
fig_compare.add_trace(go.Scatter(x=crypto_df.index, y=crypto_df['Close'], name='BTC Close', line=dict(color='#f59e0b', width=2)), row=1, col=1)
fig_compare.add_trace(go.Scatter(x=crypto_df.index, y=crypto_df['BB_Upper'], name='BTC BB Upper', line=dict(color='#f43f5e', width=1, dash='dot')), row=1, col=1)
fig_compare.add_trace(go.Scatter(x=crypto_df.index, y=crypto_df['BB_Lower'], name='BTC BB Lower', line=dict(color='#10b981', width=1, dash='dot')), row=1, col=1)

# Subplot 2: Apple Close Price + Translucent Liquidity Area
fig_compare.add_trace(go.Scatter(x=stock_df.index, y=stock_df['Close'], fill='tozeroy', fillcolor='rgba(0, 255, 255, 0.08)', name='AAPL Area', line=dict(color='#00ffff', width=2)), row=2, col=1)

fig_compare.update_layout(height=700, template='plotly_dark', title_text="🍎 vs ⚡ UNIFIED MULTI-ASSET INSIGHT TERMINAL", showlegend=True)
fig_compare.show()
```

### 🔬 2.2 Charts 3 & 4: Density Distributions & Day-of-Week Volume Analysis
We map continuous logarithmic return profiles to inspect variance tail risks alongside a categorical tracking axis showing mean volume turnover based on trading days.

```python
fig_dist = make_subplots(rows=1, cols=2, subplot_titles=('3. Return Density Distribution (BTC)', '4. Mean Volume Turnover by Weekday'))

# Density Histogram Tracking
fig_dist.add_trace(go.Histogram(x=crypto_df['Log_Return'], nbinsx=60, name='BTC Log Returns', marker_color='#a855f7', opacity=0.85), row=1, col=1)

# Day-of-Week Grouping Aggregations
weekday_volume = crypto_df.groupby('Day_Name')['Volume'].mean().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
fig_dist.add_trace(go.Bar(x=weekday_volume.index, y=weekday_volume.values, name='Mean Volume', marker=dict(color=weekday_volume.values, colorscale='Viridis')), row=1, col=2)

fig_dist.update_layout(height=500, template='plotly_dark', title_text="🔬 QUANTITATIVE VOLUME & DISTRIBUTION ANALYSIS")
fig_dist.show()
```

### 🛡️ 2.3 Charts 5 to 8: Risk Dispersion Contours & Phase Shares
Advanced diagnostic structures measuring tail variations across expansionary and contractionary macro regimes.

```python
# Charts 5 & 6: Violin Contours and Outlier Box Plots
fig_risk = make_subplots(rows=1, cols=2, subplot_titles=('5. Volatility Density Profiles', '6. Outlier Envelopes'))
fig_risk.add_trace(go.Violin(y=crypto_df['Rolling_Vol_21'], x=crypto_df['Market_State'], box_visible=True, meanline_visible=True, name='Vol Contours', line_color='#06b6d4'), row=1, col=1)
fig_risk.add_trace(go.Box(y=crypto_df['Log_Return'], x=crypto_df['Market_State'], boxpoints='outliers', name='Return Outliers', marker_color='#ec4899'), row=1, col=2)
fig_risk.update_layout(height=450, template='plotly_dark', title_text="🛡️ RISK DISPERSION REGIMES")
fig_risk.show()

# Charts 7 & 8: Categorical Market Composition Pies
fig_share = make_subplots(rows=1, cols=2, specs=[[{'type':'pie'}, {'type':'pie'}]], subplot_titles=('7. Market Phase Share', '8. Chronological Volume Weight'))
state_counts = crypto_df['Market_State'].value_counts()
fig_share.add_trace(go.Pie(labels=state_counts.index, values=state_counts.values, marker=dict(colors=['#10b981','#f43f5e'])), row=1, col=1)
day_volume = crypto_df.groupby('Day_Name')['Volume'].sum()
fig_share.add_trace(go.Pie(labels=day_volume.index, values=day_volume.values, hole=0.4), row=1, col=2)
fig_share.update_layout(height=450, template='plotly_dark', title_text="📊 CATEGORICAL MARKET SEGMENTATION")
fig_share.show()
```
