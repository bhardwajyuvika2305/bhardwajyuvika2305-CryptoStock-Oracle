### **📊 2.1 Continuous Logarithmic Returns Matrix**

### **🏹 2.2 Bounded Momentum Velocity Space (Relative Strength Index)**
#### The RSI measures momentum velocity by isolating the exponential moving average of upward price changes relative to downward price changes over an $n$-period lookback window (typically 14 trading days):
#### **$$\text{RSI} = 100 - \left( \frac{100}{1 + \text{RS}} \right), \quad \text{where } \text{RS} = \frac{\text{EMA}_n(\Delta P^+)}{\text{EMA}_n(\Delta P^-)}$$**


### **🛡️ 2.3 Dynamic Bollinger Volatility Envelopes**
#### Bollinger Bands establish dynamic boundaries around price action based on rolling standard deviations **($\sigma$)**. They adapt automatically to changing volatility regimes by expanding during high-variance periods and contracting during quiet consolidation phases:
#### **$$\text{Middle Horizon Line} = \frac{1}{n}\sum_{i=0}^{n-1} P_{t-i}$$**
##### **$$\text**{Upper Boundary} = **\text**{Middle Horizon Line} + **\left(**k \cdot \sqrt{\frac{1}{n}\sum_{i=0}^{n-1}(P_{t-i} - \text{Middle Horizon Line})^2}\right)$$**
##### **$$\text**{Lower Boundary} = **\text**{Middle Horizon Line} - **\left(**k \cdot \sqrt{\frac{1}{n}\sum_{i=0}^{n-1}(P_{t-i} - \text{Middle Horizon Line})^2}\right)$$**

```python
def execute_ingestion_pipeline(ticker, lookback_years=2):
    """
    Harvests data directly from the yfinance API layer.
    Flattens multi-index arrays and uses vectorized computations to calculate
    higher-order technical indicators in-memory.
    """
    end_date = dt.date.today()
    start_date = end_date - dt.timedelta(days=lookback_years * 365)
    
    print(f"📡 Ingesting live streaming array blocks for: {ticker}...")
    raw_data = yf.download(ticker, start=start_date, end=end_date)
    
    df = pd.DataFrame(index=raw_data.index)
    df['Open'] = raw_data['Open'].values
    df['High'] = raw_data['High'].values
    df['Low'] = raw_data['Low'].values
    df['Close'] = raw_data['Close'].values
    df['Volume'] = raw_data['Volume'].values
    
    # Mathematical Preprocessing Foundations
    df['Log_Return'] = np.log(df['Close'] / df['Close'].shift(1))
    df['Rolling_Vol_21'] = df['Log_Return'].rolling(window=21).std() * np.sqrt(252)
    df['SMA_20'] = df['Close'].rolling(20).mean()
    df['SMA_50'] = df['Close'].rolling(50).mean()
    
    # Volatility Boundaries
    df['Std_20'] = df['Close'].rolling(20).std()
    df['BB_Upper'] = df['SMA_20'] + (df['Std_20'] * 2)
    df['BB_Lower'] = df['SMA_20'] - (df['Std_20'] * 2)
    
    # Categorize Directional Registry for Advanced Segmentation Visuals
    df['Market_State'] = np.where(df['Log_Return'] >= 0, 'Expansion (Bullish)', 'Contraction (Bearish)')
    df['Day_Name'] = df.index.day_name()
    
    return df.dropna()
print("Pipeline Execution Complete. DataFrame Ready for Analysis.")
```

### **🔬 SECTION 4: HIGHER-ORDER STATISTICAL DISTRIBUTIONS & STATIONARITY TESTS**
#### ---
#### To verify that our historical trends have true structural merit rather than just showing temporary patterns, we run an **Augmented Dickey-Fuller (ADF) Test** to verify mathematical stationarity. We also compute **Skewness** and **Kurtosis** to check for till risk asymmetry

```python
def run_statistical_distribution_audit(df, asset_name):
    returns_vector = df['Log_Return'].dropna()
    
    print("\n" + "═"*80)
    print(f"📊 MATHEMATICAL ANALYSIS MATRIX REPORT FOR: {asset_name.upper()}")
    print("═"*80)
    print(f"   ▪️ Logarithmic Daily Mean:   {returns_vector.mean():.6f}")
    print(f"   ▪️ Annualized System Variance: {(returns_vector.var() * 252):.6f}")
    print(f"   ▪️ Matrix Sample Skewness:   {stats.skew(returns_vector):.4f} (Ideal Normal = 0)")
    print(f"   ▪️ Excess Sample Kurtosis:   {stats.kurtosis(returns_vector):.4f} (Ideal Normal = 0)")
    
    # Running the Augmented Dickey-Fuller (ADF) Unit Root Test
    adf_stat, p_val, _, _, crit_values, _ = adfuller(df['Close'])
    print(f"\n   🔬 Augmented Dickey-Fuller Stationarity Metrics:")
    print(f"      ▫️ ADF Unit Root Test Statistic: {adf_stat:.4f}")
    print(f"      ▫️ Calculated p-value boundary:  {p_val:.6f}")
    
    if p_val <= 0.05:
        print("\n🟩 [EVALUATION]: p-value is below alpha threshold (0.05). Reject the null hypothesis.")
        print("   The asset series is Stationary and exhibits long-term mean-reverting behavior.")
    else:
        print("\n🟨 [EVALUATION]: p-value is above alpha threshold (0.05). Fail to reject the null hypothesis.")
        print("   The asset series is Non-Stationary and displays integrated trending mechanics.")

run_statistical_distribution_audit(crypto_df, "Bitcoin Continuous Decentralized Node")
run_statistical_distribution_audit(equity_df, "Apple Corporate Equity Centralized Node")
```

# **🔮 SECTION 6: PURE MATHEMATICAL TREND HORIZON PROJECTIONS**
#### ---
### To prevent model overfitting on noisy financial data, we use a robust **Ordinary Least Squares (OLS) Linear Equation Model** to extrapolate long-term asset directions out over a 30-day projection horizon.

```python
def design_least_squares_extrapolator(df, label, projection_window=30):
    matrix = df.copy()
    matrix['Time_Index'] = np.arange(len(matrix))
    
    # Map arrays to clean baseline vectors
    X = matrix['Time_Index'].values
    Y = matrix['Close'].values
    
    slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)
    
    # Project forward values into future arrays
    future_indexes = np.arange(len(matrix), len(matrix) + projection_window)
    future_timeline = [matrix.index[-1] + dt.timedelta(days=i) for i in range(1, projection_window + 1)]
    predicted_valuations = intercept + slope * future_indexes
    
    print("\n" + "═"*80)
    print(f"🔮 OLS STRUCTURAL FORECAST LOG MATRIX: {label.upper()}")
    print("═"*80)
    print(f"   ▪️ Mathematical Trend Slope (Velocity):  {slope:.4f}")
    # Fix statistical formula rendering to clean text string formatting
    print(f"   ▪️ Statistical Confidence Metric (R2):  {(r_value**2):.4f}")
    print(f"   ▪️ Extrapolated End Target Price:       ${predicted_valuations[-1]:,.2f}")
    print("═"*80)
    
    return future_timeline, predicted_valuations

future_time, pred_vals = design_least_squares_extrapolator(crypto_df, "Bitcoin Spot Target Network")
```
